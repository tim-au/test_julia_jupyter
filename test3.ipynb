# Experimenting with PyPlot and Julia for faster plotting

First we will import relevant libraries

```julia
using PyPlot
```

### Plot 1:

```julia
θ = LinRange(0, 2π, 100)
y = sin.(θ);
```

```julia
plot(θ, y)
title(L"sin(\theta)")
xlim(0, 2π)
ylim(-1.25, 1.25)
xlabel(L"\theta")
ylabel("Amplitude");

```

### Plot 2:

```julia
x = LinRange(1, 5, 100)
y = x.^2 - 3x .- 2

plot(x, y, c = "pink")
title(L"y = x^2")
xlabel(L"x"*" mm")
ylabel(L"y"*" mm");
```

### Plot 3:

```julia
scatter(θ, y, s = 40, alpha = 0.5, c = LinRange(0, 256, length(θ)), marker = "h");
```

### Unitful

```julia
using Unitful
using LinearAlgebra

Unitful.preferunits(u"mm, s"...)
```

```julia

F = [1000, 3200, -4000, 4300, -2570, 6630]u"N"
```

```julia
r = [300u"mm", 400u"mm", 10u"inch", 15u"cm", 0.188u"m", 27u"cm"] .|> u"mm"
```

```julia
scatter(ustrip(r), ustrip(F))
title("r vs F")
xlabel("r [mm]")
ylabel("F [mm]");
```

```julia
using DataFrames

df = DataFrame(r=r, F=F)
```
